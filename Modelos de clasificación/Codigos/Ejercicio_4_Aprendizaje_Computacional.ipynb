```python
## for data
import pandas as pd
import numpy as np

## for plotting
import matplotlib.pyplot as plt
import seaborn as sns

## for statistical tests
import scipy

## for machine learning
from sklearn import model_selection, preprocessing, feature_selection, ensemble, linear_model, metrics, decomposition
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, recall_score, precision_score
from sklearn.svm import SVC
from sklearn import datasets
import matplotlib.cm as cm
```

```python
data = np.loadtxt('/content/misterious_data_2.txt')
```

```python
n_features = 324
n_samples = 234

x = np.empty((n_samples, n_features)) #data o features
y = np.empty((n_samples,), dtype=int) #target

with open('/content/misterious_data_2.txt', 'r') as file:
    for i, line in enumerate(file):
        values = line.strip().split('\t')
        y[i] = int(values[0])
        features = [float(value) for value in values[1:]]
        x[i] = features
```

**CROSS-VALIDATION**

K = 1,2,3 ... 40

```python
#Evaluación de modelos usando k-folds cross validation (k = 1, 2, 3, 4, 5 ... 40)
fin = 40;
n = 1 #valor de k para el clasificador K-NN

while n <= fin:
  n_folds = 5
  kf = StratifiedKFold(n_splits=n_folds, shuffle = True)

  acc = 0

  cv_y_test = []
  cv_y_pred = []

  for train_index, test_index in kf.split(x, y):

      # Training phase
      x_train = x[train_index, :]
      y_train = y[train_index]

      clf_cv = KNeighborsClassifier(n_neighbors = n)
      clf_cv.fit(x_train, y_train)

      # Test phase
      x_test = x[test_index, :]
      y_test = y[test_index]
      y_pred = clf_cv.predict(x_test)

      # Concatenate results of evaluation
      cv_y_test.append(y_test)
      cv_y_pred.append(y_pred)

      # Performance scores
      acc += accuracy_score(y_test, y_pred)

  # Model performance
  print("------------------Clasificador K-NN con k =", n,'------------------')
  print('-----Evaluación usando k-folds cross validation con 5 pliegues-----')
  print(classification_report(np.concatenate(cv_y_test), np.concatenate(cv_y_pred)))

  # Print average performance
  acc = acc/n_folds
  print('Exactitud del modelo: ', acc)
  n = n+1
```

```python
# SVM con kernel lineal con C = 0.000001, 0.000002 ... 0.00001
fin = 0.00001;
n = 0.000001

n_folds = 5
kf = StratifiedKFold(n_splits=n_folds, shuffle = True)

while n <= fin:

  X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
  clf = SVC(C = n, kernel = 'linear')

  acc = 0

  cv_y_test = []
  cv_y_pred = []

  for train_index, test_index in kf.split(x, y):

      # Training phase
      x_train = x[train_index, :]
      y_train = y[train_index]

      clf_cv = SVC(kernel = 'linear')
      clf_cv.fit(x_train, y_train)

      # Test phase
      x_test = x[test_index, :]
      y_test = y[test_index]
      y_pred = clf_cv.predict(x_test)

      # Concatenate results of evaluation
      cv_y_test.append(y_test)
      cv_y_pred.append(y_pred)

      # Performance scores
      acc += accuracy_score(y_test, y_pred)

  # Model performance
  print("Modelo SVM con kernel lineal:")
  print('--------------------------- C =', n,'---------------------------')
  print(classification_report(np.concatenate(cv_y_test), np.concatenate(cv_y_pred)))

  # Print average performance
  acc = acc/n_folds
  print('Exactitud del modelo: ', acc)
  n = n + 0.000001
```
